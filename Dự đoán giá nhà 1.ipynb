**TIỀN XỬ LÍ DỮ LIỆU**

```python
from sklearn.linear_model import LinearRegression
from math import ceil
```

```python
#CHUYỂN DẤU ',' THÀNH DẤU '.' TRONG CỘT BATHROOMS
df_Housing["Bathrooms"] = df_Housing["Bathrooms"].str.replace(",", ".")
df_Housing
```

```python
#ĐỔI KIỂU DỮ LIỆU CỘT BATHROOMS THÀNH FLOAT
df_Housing["Bathrooms"] = df_Housing["Bathrooms"].astype(float)
```

```python
#LÀM TRÒN DỮ LIỆU TRONG CỘT BATHROOMS
df_Housing["Bathrooms"] = df_Housing["Bathrooms"].apply(ceil)
```

```python
df_Housing
```

```python
df_Housing = df_Housing.drop("Unnamed: 0", axis = 1)
```

```python
df_Housing
```

**XÂY DỰNG MÔ HÌNH DỰ ĐOÁN**

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
X = df_Housing.iloc[:, 1:]
y = df_Housing.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.16, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn import metrics
```

```python
print("MAE: ", (metrics.mean_absolute_error(y_pred, y_test)))
print("MSE: ", (metrics.mean_squared_error(y_pred, y_test)))
print("R2 score: ", (metrics.r2_score(y_pred, y_test)))
```
